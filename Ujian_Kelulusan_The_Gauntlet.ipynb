## Langkah 2: Kode Utama untuk Analisis dan Visualisasi

Sekarang kita akan menjalankan kode utama. Sel kode di bawah ini akan melakukan semua tugas yang kita rencanakan:

1.  **Impor library** yang diperlukan (`yfinance`, `pandas`, `matplotlib`).
2.  **Tentukan saham** yang akan dianalisis (misalnya, `BBCA.JK`) dan periode waktunya.
3.  **Unduh datanya**.
4.  **Hitung SMA 20 dan SMA 50** dan tambahkan sebagai kolom baru di data kita.
5.  **Buat grafik** untuk memvisualisasikan hasilnya.

```python
# --- BAGIAN 1: MENGAMBIL DATA ---
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

ticker_saham = 'BBCA.JK'
periode = '1y'
print(f"Mengunduh data untuk {ticker_saham}...")
```

```python
# Menggunakan yfinance untuk mengunduh data saham
data = yf.download(ticker_saham, period=periode, progress=False)
print("Data berhasil diunduh.")
```

```python
# --- BAGIAN 2: MENGHITUNG INDIKATOR TEKNIKAL ---
print("Menghitung Simple Moving Averages (SMA)...")
# Menghitung SMA 20 hari pada kolom harga penutupan ('Close')
data['SMA20'] = data['Close'].rolling(window=20).mean()

# Menghitung SMA 50 hari pada kolom harga penutupan ('Close')
data['SMA50'] = data['Close'].rolling(window=50).mean()
print("Perhitungan SMA selesai.")
```

```python
print("Membuat grafik visualisasi...")
# Buat kanvas grafik dengan ukuran yang lebih besar agar mudah dibaca
plt.figure(figsize=(15, 8))

# Plot setiap garis data ke dalam grafik
plt.plot(data.index, data['Close'], label='Harga Penutupan', color='blue', alpha=0.9)
plt.plot(data.index, data['SMA20'], label='SMA 20 Hari', color='orange', linestyle='--')
plt.plot(data.index, data['SMA50'], label='SMA 50 Hari', color='red', linestyle='--')

# --- PENAMBAHAN KONTEKS PADA GRAFIK (SESUAI PERBAIKAN) ---
# Tambahkan judul yang informatif pada grafik
plt.title(f'Analisis Harga Saham {ticker_saham} dengan Simple Moving Average', fontsize=18)

# Beri label pada sumbu X (Tanggal)
plt.xlabel('Tanggal', fontsize=12)

# Beri label pada sumbu Y (Harga)
plt.ylabel('Harga Penutupan (IDR)', fontsize=12)

# Tampilkan legenda untuk menjelaskan setiap garis plot
plt.legend(fontsize=10)

# Tambahkan grid untuk membantu pembacaan nilai pada grafik
plt.grid(True)

# Tampilkan grafik yang sudah jadi
plt.show()
print("Proses selesai. Grafik telah ditampilkan.")
```

## Analisis Hasil

Grafik di atas menunjukkan pergerakan harga penutupan saham **BBCA.JK** (garis biru solid) selama satu tahun terakhir.

- **Garis Oranye (SMA 20)**: Menunjukkan tren harga jangka pendek.
- **Garis Merah (SMA 50)**: Menunjukkan tren harga jangka menengah/panjang.

Perpotongan antara kedua garis SMA ini sering digunakan sebagai sinyal oleh para analis teknikal:
- **Golden Cross**: Jika garis SMA 20 memotong ke **atas** garis SMA 50, ini bisa dianggap sebagai sinyal tren naik (bullish).
- **Death Cross**: Jika garis SMA 20 memotong ke **bawah** garis SMA 50, ini bisa dianggap sebagai sinyal tren turun (bearish).
